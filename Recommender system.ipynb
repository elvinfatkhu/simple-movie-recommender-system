```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from ast import literal_eval

pd.options.display.max_columns = 999
```

```python
df = pd.read_csv('tmdb_5000_credits.csv')
```

```python
df2 = pd.read_csv('tmdb_5000_movies.csv')
```

```python
df['genres'] = df2['genres']
```

```python
df['keywords'] = df2['keywords']
```

```python
df.drop(columns = 'movie_id', inplace = True)
```

```python
df
```

```python
features=['cast','crew','genres', 'keywords']
for feature in features:
    df[feature] =df[feature].apply(literal_eval) ## parse the stingfiled features into python object
```

```python
df
```

```python
def clean_directors(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
df['director'] = df['crew'].apply(clean_directors)
```

```python
def get_list(x):
    names = []
    for i in x:
        names.append(i['name'])
    if len(names) > 3:
        names = names[:3]
        return names
    else:
        return names
        



```

```python
df['genres2'] = df['genres'].apply(get_list)
```

```python
df['cast2'] = df['cast'].apply(get_list)
```

```python
df['crew'] = df['crew'].apply(get_list)
```

```python
df['keywords'] = df['keywords'].apply(get_list)
```

```python
df.drop(columns = ['cast' ,'crew', 'genres'], inplace = True)
```

```python
df
```

```python
df.columns = ['title', 'keywords', 'director', 'genres', 'cast']
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ","")) for i in x]
    else:
        if isinstance(x,str):
            return str.lower(x.replace(" ",""))
        else:
            return ""
```

```python
df['genres'] = df['genres'].apply(clean_data)
```

```python
df['cast'] = df['cast'].apply(clean_data)
```

```python
df['director'] = df['director'].apply(clean_data)
```

```python
def combo(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' '+ ' '.join(x['director']) + ' ' + ' '.join(x['genres'])
```

```python
df['allfeature'] = df.apply(combo, axis = 1)
```

```python
CV = CountVectorizer()
```

```python
movie_matrix = CV.fit_transform(df['allfeature'])
```

```python
cos_score =  cosine_similarity(movie_matrix)
```

```python
df = df.reset_index()
```

```python
def findindex(movie):
    return list(df[df['title'] == movie]['index'])[0]
```

```python
def findmovie(index):
    return list(df[df['index'] == index]['title'])[0]
```

```python
movie_name = []
for i in df['title']:
    movie_name.append(i)
```

```python
watched_movie = input('Input name of movie: ').title()
if watched_movie not in movie_name:
    print('{} is not in our database'.format(watched_movie))
else:
    index_watched = findindex(watched_movie)
    similar_movie = list(enumerate(cos_score[index_watched]))
    sorted_movies = sorted(similar_movie, key=lambda x: x[1], reverse=True)[1:]
    print(f"Top 7 Movies Recommendation Based on your Watched Movie : {watched_movie}")
    print("="*50)
    count = 1
    for i in sorted_movies:
        print(findmovie(i[0]))
        count += 1
        if count == 7:
            break
```
